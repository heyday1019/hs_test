```python
import random

# 필드명 리스트
fields = ["도서명", "저자", "대출건수"]

# 도서명 샘플 리스트
book_titles = ["인공지능과 흙", "데이터 사이언스의 미래", "파이썬 프로그래밍", "딥러닝 기초", "자연어 처리와 AI", 
               "머신러닝 실전", "빅데이터 분석", "통계학 입문", "고급 프로그래밍", "알고리즘 디자인"]

# 저자 샘플 리스트
authors = ["김동훈 지음", "이영희 지음", "박철수 지음", "최영준 지음", "유재석 지음", 
           "정형돈 지음", "하하 지음", "노홍철 지음", "김태호 지음", "양세형 지음"]

# 샘플 데이터 생성 함수
def generate_sample_data(num_samples=100):
    samples = []
    for _ in range(num_samples):
        book_title = random.choice(book_titles)
        author = random.choice(authors)
        loan_count = random.randint(0, 100)
        samples.append((book_title, author, loan_count))
    return samples

# 100개의 샘플 데이터 생성
sample_data = generate_sample_data(100)

# 출력
for data in sample_data:
    print(f"{fields[0]}: {data[0]}, {fields[1]}: {data[1]}, {fields[2]}: {data[2]}")
```
