# Matplotlib 기본

## Figure 객체
### figure :
- matplotlib 의 그래프 요소를 모두 담고 있는 최상위 객체
- matplotlib으로 그래프를 그릴 때 자동으로 figure가 생성되고 그래프가 그려진 후 삭제된다.
- figure 객체를 생성하면 다양한 옵션을 컨트롤 할 수 있다

```python
import matplotlib.pyplot as plt # .pyplot은서브 모듈로, 간단한 코드로 다양한 그래프와 도표를 그릴 수 있게 해줍니다   
fig, ax = plt.subplots(3, 2, figsize=(6,6)) # figsize 매개변수 - 그래프의 크기를 튜플로 지정, 단위는 인치
ax[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16]) # .plot : 라인 그래프 (Line Plot)
ax[0, 1].scatter([1, 2, 3, 4], [1, 4, 9, 16]) # .scatter : 산점도 (Scatter Plot)
ax[1, 0].bar([1, 2, 3, 4], [1, 4, 9, 16]) # 막대 그래프 (Bar Plot)
ax[1, 1].hist([1, 1, 2, 2, 2, 3, 4, 4, 5]) # 히스토그램 (Histogram)
ax[2, 0].plot([1, 2, 3, 4], [1, 10, 100, 1000])
ax[2, 0].set_yscale('log')
plt.show()
```
