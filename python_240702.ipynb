```python
import seaborn as sns 

tips = sns.load_dataset("tips")
tips.head()
```

# GroupBy
- 데이터셋.groupby(컬럼명).집계함수
  + 해당 컬럼은 범주 형태
- 포인트 1.데이터 형태가 바뀜
  + 여러분들이 추가적인 조작

```python
# 남자와 여자 누가 팁을  평균적으로 더 많이 줄까요/
tips.head(1)
tips.groupby('sex', observed=True)['tip'].mean()
```

```python
tips.groupby('sex', observed=True)['tip'].mean().reset_index()
```

```python
tips.groupby(['sex', 'day', 'smoker'], observed=True)['tip'].mean().reset_index()
```

```python
tips.groupby(['sex', 'day'], observed=True)['tip'].agg(['mean', 'std', 'max', 'min']).reset_index()
```

```python
tips.groupby(['sex', 'day'], observed=True).size()
```

```python
tips.groupby(['sex', 'day'], observed=True).size().reset_index() # .reset_index()추가
```
